```python
# Inheritance in Python OOPs 
# Single inheritance example
class Car:
    def __init__(self, windows,doors, engineType):
        self.windows=windows
        self.doors=doors
        self.engineType=engineType
    def drive(self):
        return f"The {self.engineType} car is being driven"
    

class Tesla(Car):
    def __init__(self, windows, doors, engineType, batteryCapacity):
        super().__init__(windows, doors, engineType)
        self.batteryCapacity=batteryCapacity
    def lowBattery(self):
        return f"Low battery mode of {self.batteryCapacity} kWh activated"
```

```python
car1=Car(4,4,'Petrol')
print(car1.drive())
```

```python
modelX=Tesla(4, 4,'Electric', 1000)
print(modelX.drive())
print(modelX.lowBattery())
```

```python
# Multiple Inheritance example 

# Base class 1
class Animal:
    def __init__(self, name):
        self.name=name

    def speak(self):
        return f"{self.name} makes a sound"
    
# Base class 2
class Pet:
    def __init__(self, owner):
        self.owner=owner

# Derived class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} barks"
```

```python
dog=Dog("Buddy", "Alice")
print(dog.speak())  # Output: Buddy barks
print(f"Owner of the dog is {dog.owner}")  # Output: Owner of the dog is Alice
```
