### Pandas-> Series and DataFrame:

Pandas is a powerful data manipultion libraty in python, widely used for data analysis and data cleaning. It provides two primary data structures

    - Series -> 1D array like object
    
    - DataFrame -> 2D size-mutable and potentially heterogeneous tabular data structure with labeled axes (rows and columns).



```python
import numpy as np
import pandas as pd
```

```python
data=[1,2,3,4,5]
series=pd.Series(data)
print(type(series))
print(series)
```

```python
# Create a series from a dictionary
data_dict = {'a': 10, 'b': 20, 'c': 30}
series_dict = pd.Series(data_dict)
print(type(series_dict))
print("\nSeries from dictionary:")
print(series_dict)
```

```python
data=[10,20,30,40,50]
index_labels=['1.0','1.1','1.2','1.3','1.4']
series_with_index = pd.Series(data, index=index_labels)
print("\nSeries with custom index:")
print(series_with_index)
```

```python
# Dataframes
# Create a DataFrame from a dictionary of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)
print(type(df))
print(df)
```

```python
# Demonstrating differences between loc and iloc

# Create a simple DataFrame for clear examples
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['NYC', 'LA', 'Chicago', 'Miami', 'Seattle']
}
df_demo = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])
print("Demo DataFrame:")
print(df_demo)
print("\n" + "="*50)

# 1. Accessing single row
print("\n1. SINGLE ROW ACCESS:")
print("df.loc['B']:  # Using label 'B'")
print(df_demo.loc['B'])
print("\ndf.iloc[1]:  # Using position 1 (second row)")
print(df_demo.iloc[1])
print("\n" + "="*50)

# 2. Slicing rows
print("\n2. SLICING ROWS:")
print("df.loc['B':'D']:  # Labels B to D (inclusive)")
print(df_demo.loc['B':'D'])
print("\ndf.iloc[1:4]:  # Positions 1 to 4 (exclusive of 4)")
print(df_demo.iloc[1:4])
print("\n" + "="*50)

# 3. Accessing specific cell
print("\n3. SPECIFIC CELL ACCESS:")
print("df.loc['C', 'Age']:  # Row label 'C', column 'Age'")
print(df_demo.loc['C', 'Age'])
print("\ndf.iloc[2, 1]:  # Row position 2, column position 1")
print(df_demo.iloc[2, 1])
print("\n" + "="*50)

# 4. Multiple columns
print("\n4. MULTIPLE COLUMNS:")
print("df.loc['A':'C', ['Name', 'City']]:  # Labels with column names")
print(df_demo.loc['A':'C', ['Name', 'City']])
print("\ndf.iloc[0:3, [0, 2]]:  # Positions with column positions")
print(df_demo.iloc[0:3, [0, 2]])
```

```python
# 5. Boolean indexing with loc (iloc doesn't support this directly)
print("\n5. BOOLEAN INDEXING:")
print("df.loc[df['Age'] > 30]:  # Filter rows where Age > 30")
print(df_demo.loc[df_demo['Age'] > 30])

# Key takeaways
print("\n" + "="*60)
print("KEY DIFFERENCES SUMMARY:")
print(" loc[]   Uses LABELS (index/column names)")
print(" iloc[]  Uses INTEGER POSITIONS (0-based)")
print(" loc[]   Includes both endpoints in slicing")
print(" iloc[]  Excludes end point in slicing (like Python lists)")
print(" loc[]   Supports boolean indexing")
print(" iloc[]  Only works with integer positions")
print("="*60)
```
