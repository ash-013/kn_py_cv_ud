## Magic Methods

Magic methods in python are also known as dunder methods (fouble underscore methods __method_name__), are special methods that start and end with double underscores. These methods enable us to define the behavior of objects for built in operators such as arithematic operators,  comparison operators etc...

## Magic Methods (more)

Magic methods are predefined methods in python that we can override to change the behavior of objects. Some common magic mathods include: 
- init ->            initialize a new instance of a class
- str ->         returns a string representation of an object
- repr ->            returns a official string representation of an object
- len ->         returns the length of an object
- getitem ->         gets an item form a container
- setitem ->         sets an item in a container

```python
# Basic magic methods:

class Person:
    def __init__(self, name, age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return f'{self.name}, {self.age} years old'
    
    def __repr__(self):
        return f'Person, name is {self.name} and age is {self.age}'

person=Person('Holstein', 12)
print(person)
print(repr(person))
```
