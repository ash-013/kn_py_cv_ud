```python
# Instance variable and methods
class Dog:
    def __init__(self, name, age):
        self.legs=4
        self.ears=2
        self.tail=1
        self.name=name
        self.age=age
    
    def bark(self):
        return "Woof! Woof!"

# Creating objects
dog1=Dog("Tommy", 5)
dog2=Dog("Bruno", 3)

print(f'{dog1.name} has {dog1.legs} legs, {dog1.ears} ears, and {dog1.tail} tail.')
print(f'{dog2.name} has {dog2.legs} legs, {dog2.ears} ears, and {dog2.tail} tail.')

print(f'{dog1.name} says: {dog1.bark()}')
print(f'{dog2.name} says: {dog2.bark()}')
```

```python
# Define a class with instance methods

class Cat:
    def __init__(self, name, age):
        self.name=name
        self.age=age

    def meow(self):
        return f'{self.name} says Meow! Meow!'
    
cat1=Cat("Whiskers", 2)
cat2=Cat("Mittens", 4)


print(f'{cat1.meow()}')
print(f'{cat2.meow()}')
```

```python
# Modeling a Bank Account
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self, amount):
        self.balance=self.balance+amount
        return f'Deposit successful\nNew balance is: {self.balance}'
    
    def withdraw(self, amount):
        if(amount<self.balance):
            self.balance=self.balance-amount
            return f'Withdraw successful!\nNew balance is: {self.balance}'
        else:
            print('Insufficent funds!')

    def checkBalance(self):
        return f'Account balance is :{self.balance}'
```

```python
ajeet=BankAccount('Ajeet Bharti', 4000)
print(ajeet.deposit(200))
print(ajeet.withdraw(3000))
print(ajeet.checkBalance())
```
