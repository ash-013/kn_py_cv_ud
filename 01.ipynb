```python
# Instance Variables and Instance Methods
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

dog1=Dog("Buddy", 3)
dog2=Dog("Max", 5)

print(dog1.name)
print(dog2.age)
```

```python
# Define class with instance methods
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return "Woof!"

dog1=Dog("Buddy", 3)
dog2=Dog("Max", 5)

print(f'{dog1.name} says {dog1.bark()}')
print(f'{dog2.name} says {dog2.bark()}')
```

```python
# Modeling a BANK ACCOUNT
class BankAccount:
    def __init__(self, account_holder_name, balance=0):
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance
    
    def withdraw(self, amount):
        if amount > self.balance:
            return "Insufficient funds"
        self.balance -= amount
        return self.balance
    
    def get_balance(self):
        return self.balance
```

```python
owner1=BankAccount("Alice", 1000)
owner2=BankAccount("Bob", 200)

print(f"{owner1.account_holder_name} initial balance: {owner1.balance}")
print(f"{owner2.account_holder_name} initial balance: {owner2.balance}")
```

```python
owner1.deposit(500)
print(f"{owner1.account_holder_name} balance after deposit: {owner1.balance}")
owner2.withdraw(50)
print(f"{owner2.account_holder_name} balance after withdrawal: {owner2.balance}")
```

```python
print(f"Alice's balance: {owner1.get_balance()}")
print(f"Bob's balance: {owner2.get_balance()}")
```
