#### Custom exception (Raise and Throw an exception)

```python
class Error(Exception):
    pass

class DoBException(Error):
    pass
```

```python
year=int(input("Enter your birth year: "))
age=2026-year

try:
    if year<=30 and age>=20:
        print("The age is valid")
    else:
        raise DoBException
        
except DoBException:
    print("The age is not valid, enter a valid year of birth")
```
