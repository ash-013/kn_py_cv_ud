```python
class Animal:
    def __init__(self, species):
        self.species=species

class Cow():
    def __init__(self, cowName, breed):
        self.cowName=cowName
        self.breed=breed

class Owner(Cow, Animal): 
    def __init__(self, name, age):
        Animal.__init__(self, 'Cow')
        Cow.__init__(self, 'Cattle', breed'Holstein')  
        self.name=name
        self.age=age

ajeet=Owner('Ajeet Bharti', 24)
print(ajeet.name)
print(ajeet.age)
print(ajeet.species) 
print(ajeet.breed)   
```
