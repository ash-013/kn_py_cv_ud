### Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing the programming complexity and effort. 

```python
from abc import ABC, abstractmethod

# Abstract Base Class
class Vehicle(ABC):
    def drive(self):
        return 'The vehicle is used for driving'

    @abstractmethod
    def start_engine(self):
        pass

# Derived Class
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"
    
def operate_vehicle(vehicle):
    return vehicle.start_engine(), vehicle.drive()
```

```python
car=Car()
print(operate_vehicle(car))
```
