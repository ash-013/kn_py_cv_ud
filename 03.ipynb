```python
# Base class
class Animal:
    def speak(self):
        return "Animal speaks"

# Derived class    
class Dog(Animal):
    def speak(self):
        return "Woof! Woof!"
    
# Derived class    
class Cat(Animal):
    def speak(self):
        return "Meow! Meow!"
    
# Function to demonstrate polymorphism
def animal_speak(animal):
    print(animal.speak())
    
cat = Cat()
dog = Dog()

animal_speak(cat)
animal_speak(dog)
```

```python
# Polymorphism with functions and methods
class Shape:
    def area(self):
        return "Area not defined"

class Rectangle(Shape):
    def __init__(self, width, height, shape_type="Rectangle"):
        self.width = width
        self.height = height
        self.shape_type = shape_type
        
    def area(self):
        return self.width * self.height
    

class Circle(Shape):
    def __init__(self, radius, shape_type="Circle"):
        self.radius = radius
        self.radius = radius
        self.shape_type = shape_type
        
    def area(self):
        return 3.14 * self.radius * self.radius

rectangle=Rectangle(4, 5)
circle=Circle(3)

print(f"Area of {rectangle.shape_type} is: {rectangle.area()}")
print(f"Area of {circle.shape_type} is: {circle.area()}")
```

```python
# Polymorphism with abstract base classes

from abc import ABC, abstractmethod
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass
    
    @abstractmethod
    def stop_engine(self):
        return "Engine stopped."

# Derived class 1
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started with a key."
    def stop_engine(self):
        return 'Car has been stopped'
    
# Derived class 2
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started with a button."
    def stop_engine(self):
        return 'Motorcycle has been stopped'
    
car = Car()
motorcycle = Motorcycle()

print(car.start_engine())
print(car.stop_engine())
print(motorcycle.start_engine())
print(motorcycle.stop_engine())
```
