```python
# Inheritance
class Car:
    def __init__(self, windows, doors, engine_type):
        self.windows = windows
        self.doors = doors
        self.engine_type = engine_type

    def drive(self):
        print(f"The person will drive the {self.engine_type} car.")
```

```python
car1=Car(4,4,"Petrol")
car1.drive()
```

```python
# Single inheritance
class Tesla(Car):
    def __init__(self, windows, doors, engine_type, battery_capacity):
        super().__init__(windows, doors, engine_type)
        self.battery_capacity = battery_capacity

    def drive(self):
        print(f"The person will drive the electric Tesla with a battery capacity of {self.battery_capacity} kWh.")

    def self_driving(self):
        print("The Tesla is driving itself.")
```

```python
ev=Tesla(4,4,"Electric",100)
ev.self_driving()
ev.drive()
```

```python
# Multiple Inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    
    def walk(self):
        print(f'The animal is walking.')

class Pet:
    def __init__(self, owner):
        self.owner = owner
    
    def play(self):
        print(f'The pet is playing with its owner {self.owner}.')

class Dog(Animal, Pet):
    def __init__(self, name, owner, owner_address):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        self.owner_address = owner_address
    
    def identity(self):
        print(f'{self.name} the dog, the owner is {self.owner} living at {self.owner_address}.')
```

```python
fluffy=Dog("Fluffy", "Alice", "Chennai")
fluffy.identity()
```
