### Iterators

Iterators are advanced Python concepts that all for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing a underlying structure.

```python
my_list=[1,2,3,3,4,5]
for i in my_list:
    print(i, end=" ")
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
# Iterables in Python are objects that can be iterated upon, meaning you can traverse through all the values. Examples of iterables include lists, tuples, strings, and dictionaries.

iterator=iter(my_list)
print(type(iterator))
```

```python
# Iterate through the iterable list using __next__()
print(iterator.__next__())


# Error when the iterator reaches the end of the iterable
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
print(iterator.__next__())
# print(iterator.__next__())  # This will raise StopIteration error
```

```python
# next(iterator)  # -> Also works like __next__()
```

```python
iterator1=iter(my_list)
```

```python
try: 
    print(next(iterator1))
except StopIteration:
    print("End of the iterator reached.")
```

```python
# String Iteration
my_string="Hello, World!"
string_iterator=iter(my_string)

print(next(string_iterator))  # H
print(string_iterator.__next__())  # e
print(next(string_iterator))  # l
print(string_iterator.__next__())  # l
print(next(string_iterator))  # o
print(string_iterator.__next__())  # ,
print(next(string_iterator))  #   ' '
print(string_iterator.__next__())  # W
print(next(string_iterator))  # o
print(string_iterator.__next__())  # r
print(next(string_iterator))  # l
print(string_iterator.__next__())  # d
print(next(string_iterator))  # !
# print(string_iterator.__next__())  # This will raise StopIteration error
```
